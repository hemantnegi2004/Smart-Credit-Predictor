```python
df = pd.read_csv('test_file.csv')
```

```python
df.head(3)
```

```python
df.drop(columns='cc_cons',inplace=True)
```

## Handling Null Values

```python
for col in df.columns:
    print(col,':',df[col].isnull().sum())
```

## Converting categorical data to numerical data

```python
mapping_df = pd.read_csv('mapping_dict_capstoneproject.csv')
mapping_dict = dict(mapping_df.values)
```

```python
mapping_dict
```

```python
df = df[(df['gender']!='0')&(df['Income']!='0')]
```

```python
cat_col = ['loan_enq', 'account_type', 'gender', 'Income']
for col in cat_col:
    df[col] = df[col].apply(lambda x: mapping_dict[x])
```

```python
df.head(3)
```

## Standardization

```python
import pickle
with open('./standard_scaler.pkl','rb') as file:
    ss = pickle.load(file)
```

```python
final_df = df['ID']
df.head()
```

```python
x_ft = ss.fit_transform(df)
x_ft = pd.DataFrame(x_ft)
x_ft.head(3)
```

## Predicting Final Data

```python
with open('./best_model.pkl','rb') as file:
    lr = pickle.load(file)
```

```python
lr
```

```python
pred = lr.predict(x_ft)
pred_df = pd.DataFrame(pred,columns=['cc_cons'])
pred_df.head(3)
```

```python
submit_df = pd.concat([final_df,pred_df],axis=1)
submit_df.head(3)
```

```python
submit_df.to_csv('Final_submission.csv',index=False)
```
